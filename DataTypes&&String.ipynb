**DATA TYPES:**
**NUMERIC**

```python
import sys
val1=10 # Integer Data Type
print(val1)
print(type(val1)) # type of object
print(sys.getsizeof(val1)) # size of integer object in bytes
print(val1,'is Integer?',isinstance(val1,int)) #isintance cheking the value is integer
```

```python
val2=91.3 #float data type
print(val2)
print(type(val2))
print(sys.getsizeof(val2))
print(val2,'is Float?',isinstance(val2,float))
```
